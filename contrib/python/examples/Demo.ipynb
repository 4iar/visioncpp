# VisionCpp Python bindings

This Jupyter notebook provides examples of using VisionCpp for GPU-accelerated python programming.

```python
# preamble
%matplotlib inline  
import visioncpp as vp
```

Initialize VisionCpp to your [ComputeCpp](https://www.codeplay.com/products/computesuite/computecpp) installation:

```python
vp.init("~/ComputeCpp")
```

**Loading data:**

Load the test image using an `Image` node:

```python
img = vp.Image("./lena.jpg")
img
```

**Displaying results:**

Display the results of an expression using a `show` node:

```python
out = vp.show(img)
out
```

**Executing expressions:**

Use `vp.run()` to evaluate an expression tree:

```python
vp.run(out)
```

**Example expression tree:**

```python
bgr2rgb = vp.BGRToRGB(img)
out = vp.show(bgr2rgb)
out
```

```python
vp.run(out)
```

**Example expression tree:**

```python
node1 = vp.BGRToRGB(img)
node2 = vp.U8C3ToF32C3(node1)
node3 = vp.RGBToHSV(node2)
node4 = vp.HSVToRGB(node3)
node5 = vp.F32C3ToU8C3(node4)
node6 = vp.RGBToBGR(node5)
out = vp.show(node6)
out
```

```python
vp.run(out)
```
